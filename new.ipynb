```python
import sounddevice as sd
import soundfile as sf

fs = 16000
seconds = 6

print("Speak now...")
audio = sd.rec(int(seconds * fs), samplerate=fs, channels=1)
sd.wait()

sf.write("input.wav", audio, fs)
print("Audio recorded and saved as input.wav")
```

```python
import whisper

model = whisper.load_model("small")
result = model.transcribe("input.wav", language="ne")

text = result["text"]
print(text)
with open("output.txt", "w") as f:
    f.write(text)
print("Transcription saved as output.txt")
```

```python
normalization_map = {
    "बिजुलि": "बिजुली",
    "नागरीकता": "नागरिकता"
}

def normalize(text):
    for wrong, correct in normalization_map.items():
        text = text.replace(wrong, correct)
    return text

text = normalize(text)
```

**Reasoning**:
Now that `gensim` is installed, I will import the necessary `api` module, load a pre-trained Word2Vec model, and then demonstrate retrieving word vectors and finding similar words for 'social' and 'service' as instructed.



```python
import gensim.downloader as api

print("Downloading word2vec-google-news-300 model (this may take a while)...")
word_vectors = api.load("word2vec-google-news-300")
print("Model downloaded successfully.")

# Access and print word vectors for 'social' and 'service'
print("\nWord vector for 'social':")
print(word_vectors['social'])

print("\nWord vector for 'service':")
print(word_vectors['service'])

# Find and display words most similar to 'social' and 'service'
print("\nWords most similar to 'social':")
print(word_vectors.most_similar('social'))

print("\nWords most similar to 'service':")
print(word_vectors.most_similar('service'))
```
