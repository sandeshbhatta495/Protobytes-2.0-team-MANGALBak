```python
!apt-get install -y portaudio19-dev
!pip install sounddevice
import sounddevice as sd
import soundfile as sf

fs = 16000
seconds = 6

print("Speak now...")
audio = sd.rec(int(seconds * fs), samplerate=fs, channels=1)
sd.wait()

sf.write("input.wav", audio, fs)
print("Audio recorded and saved as input.wav")
```

```python
!pip install openai-whisper
import whisper

model = whisper.load_model("small")
result = model.transcribe("input.wav", language="ne")

text = result["text"]
print(text)
with open("output.txt", "w") as f:
    f.write(text)
print("Transcription saved as output.txt")
```

```python
normalization_map = {
    "बिजुलि": "बिजुली",
    "नागरीकता": "नागरिकता"
}

def normalize(text):
    for wrong, correct in normalization_map.items():
        text = text.replace(wrong, correct)
    return text

text = normalize(text)
```
